<h2>What is It?</h2>
<ul>
    <li>Kinematic calibration adjusts the kinematic model.</li>
    <li>Corrects differences between theoretical model and actual robot.</li>
    <li>Increases accuracy of robot.</li>
    <li>Does not increase repeatability (usually determined by hardware).</li>
</ul>

<a id="basics-of-quaternions"></a>
<h1  style="font-size: 4.5em; border-bottom: 2px solid #ddd; padding-bottom: 10px;" >Quaternions Overview</h1>
<p>Quaternions extend <strong>complex numbers</strong> and are pivotal in Robotics for describing rotations.</p>

<h2>Unit Quaternions:</h2>
<ul>
    <li>Representation: $ q = w + xi + yj + zk $ with real numbers $ w, x, y, z $ and quaternion units $ i, j, k $.</li>
    <li>Unit quaternions have the magnitude $ \|q\| = 1$.</li>
    <li>Unit quaternions correspond to points on the $S^3$ unit hypersphere, representing unique 3D rotations.</li>
    <li>The $S^3$ hypersphere, is a higher-dimensional analogue of the familiar 2-sphere (the surface of a ball in three-dimensional space).</li>
    <li>Quaternion Interpolation (SLERP) uses the shortest path on the \(S^3\) sphere.</li>
    <li>Quaternions are more computationally efficient than rotation matrices.</li>
    <li>They provide a compact representation of 3D rotations using only four scalar values.</li>
</ul>
<h2>Pure Quaternions:</h2>
<ul>
    <li>Representation: $ \bar{q} = 0 + xi + yj + zk $.</li>
    <li>real part of the quaternion $w$ is zero.</li>
    <li>represent vectors in 3D space.</li>
    <li>used to represent quaternion logarithm and to rotate 3D vectors.</li>
</ul>
</div>

````python
# code to visualize Quaternions
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
from DualQuaternions.Quaternion import Quaternion
from ipywidgets import widgets
from matplotlib.colors import LightSource
from mpl_toolkits.mplot3d import Axes3D


def create_3d_plot():
    
    plt.ioff()
    
    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111, projection='3d')
    fig.canvas.header_visible = False
    fig.canvas.layout.min_height = '400px'
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.set_axis_off()
    ax.set_box_aspect([1, 1, 1])
    ax.set_facecolor('white')
    
    for spine in ax.spines.values():
        spine.set_visible(False)

    plt.tight_layout()
    
    start_point = [0, 0, 0]
    R_base = np.eye(3)*1.5
    draw_frame(ax, start_point, R_base)
    
    return fig, ax

def draw_frame(ax, start_point, R):
    
    x_axis = ax.quiver(*start_point, *R[:,0], arrow_length_ratio = 0.1, linewidth = 1, color='r')
    y_axis = ax.quiver(*start_point, *R[:,1], arrow_length_ratio = 0.1, linewidth = 1, color='g')
    z_axis = ax.quiver(*start_point, *R[:,2], arrow_length_ratio = 0.1, linewidth = 1, color='b')
    return x_axis, y_axis, z_axis

def create_angle_slider(name):
    slider_width = '70%'

    slider = widgets.FloatSlider(orientation='horizontal',description=name, value=0.0, min=-2*np.pi, max=2*np.pi, step = 0.01, layout={'width': slider_width})
    return slider
  
def create_quaternion_sliders():
    slider_width = '70%'

    angle_slider = widgets.FloatSlider(orientation='horizontal',description='angle:', value=0.0, min=-2*np.pi, max=2*np.pi, step = 0.01, layout={'width': slider_width})
    azimuth_slider = widgets.FloatSlider( orientation='horizontal', description='Rotationaxis azimuth:', value=0.0, min=-np.pi, max=np.pi, step=0.01, layout={'width': slider_width})
    elevation_slider = widgets.FloatSlider(orientation='horizontal', description='Rotationaxis elevation:', value=0.0, min=-np.pi/2, max=np.pi/2, step=0.01, layout={'width': slider_width})
    
    return angle_slider, azimuth_slider, elevation_slider

def create_position_sliders():
    slider_width = '70%'

    x_slider = widgets.FloatSlider(orientation='horizontal',description='x pos:', value=0.0, min=-1, max=1, step = 0.01, layout={'width': slider_width})
    y_slider = widgets.FloatSlider( orientation='horizontal', description='y pos:', value=0.0, min=-1, max=1, step=0.01, layout={'width': slider_width})
    z_slider = widgets.FloatSlider(orientation='horizontal', description='z pos:', value=0.0, min=-1, max=1, step=0.01, layout={'width': slider_width})
    
    return x_slider, y_slider, z_slider

def spherical_coordinates(azimuth, elevation):
    x = np.sin(np.pi/2 -elevation) * np.cos(azimuth)
    y = np.sin(np.pi/2 -elevation) * np.sin(azimuth)
    z = np.cos(np.pi/2 -elevation)
    vector = np.array([x, y, z])
    return vector

def create_textbox(name):
    slider_width = '70%'
    display = widgets.Text(description=name, value='', layout={'width': slider_width})
    return display

def create_quiver(ax, start_point, direction, c):
    quiver = ax.quiver(*start_point, *direction, arrow_length_ratio=0.1, linewidth = 1, color = c)
    return quiver
    
def update_arc(arc, quaternion, p):
    t_values = np.linspace(0, 1, 50)
    arc_points = np.array([(Quaternion.slerp_old(Quaternion(1,0,0,0), quaternion, t)*p*Quaternion.slerp_old(Quaternion(1,0,0,0), quaternion, t).inverse()).getVector().flatten() for t in t_values])

    arc.set_data(arc_points[:,0], arc_points[:,1])
    arc.set_3d_properties(arc_points[:,2])
    
def draw_sphere(ax, r):
    u = np.linspace(0, 2 * np.pi, 20)
    v = np.linspace(0, np.pi, 20)

    light = LightSource(azdeg=0, altdeg=45)
    x = r * np.outer(np.cos(u), np.sin(v))
    y = r * np.outer(np.sin(u), np.sin(v))
    z = r * np.outer(np.ones(np.size(u)), np.cos(v))
    shaded = light.shade(z, cmap=plt.cm.Greys, vert_exag=1, blend_mode='soft')
    sphere = ax.plot_surface(x, y, z, facecolors=shaded, linewidth=0, antialiased=True, alpha=0.05)
    sphere_wire = ax.plot_wireframe(x, y, z, color='k', linewidth=0.4, alpha=0.1)
    return sphere, sphere_wire
````

## Axis-Angle Constructor:

The axis-angle representation of a rotation specifies an axis (a unit vector) and an angle of rotation about that axis. This representation can be converted to a quaternion.

Given an axis $ \hat{n} = [n_x, n_y, n_z] $ and an angle $ \theta $, the corresponding quaternion $ q $ is given by:

$$ q = \cos\left(\frac{\theta}{2}\right) + \sin\left(\frac{\theta}{2}\right)\hat{n} $$

---
#### Python code from Quaternion class:

```python
@classmethod
def fromAxisAngle(cls, theta, vec):
    w = np.cos(theta*0.5)
    v = vec*np.sin(theta*0.5);
    return cls(w, v[0], v[1], v[2])

```

````python
from DualQuaternions.Quaternion import Quaternion

# y axis as rotation axis
n = np.array([0,1,0])

theta = np.pi/4 # unit: rad

q = Quaternion.fromAxisAngle(theta, n)

print(q)
print(q.asRotationMatrix())
````

<div style="font-family: Arial, sans-serif;">
<h2>Quaternion Exponential:</h2>
    
<h4>Recap: Euler's Formula for Complex Numbers:</h4>
<p>
    Euler's formula establishes a deep relationship between the exponential function and the trigonometric functions for complex numbers. It is given by:
</p>
    
<p style="text-align: left; padding: 10px; font-size: 1.2em;">
    $$ e^{\theta i} = \cos(\theta) + i \sin(\theta) $$
</p>
    
<p>
    Given a pure quaternion $q = 0 + v$, the exponential of a quaternion is defined similarly to the exponential of a complex number:
<p style="text-align: left; padding: 10px; font-size: 1.2em;">
   $$ e^{q} = e^{\frac{\theta}{2}\hat{n}} = \cos(\|v\|) + \frac{v}{\|v\|} \sin(\|v\|)$$
</p>
Note:
<ul>
    <li>the scalar part $w$ of the quaternion log is 0 for unit quaternions.</li>
    <li>$\|v\| = \frac{\theta}{2}$</li>
    <li>zero-angle singularity $\|v\| = 0$ (can be fixed with tailor series).</li>
    <li>note similarity to axis angle contructor.</li>
</ul>

</div>

---
#### Python code from Quaternion class:

```python
@classmethod
def exp(cls, q):
    theta_half = np.linalg.norm(q.getVector())

    frac_sin_theta_half = 0

    if theta_half < 0.001:
        frac_sin_theta_half = 1 - theta_half**2/6.0 + theta_half**4/120.0
    else:
        frac_sin_theta_half = np.sin(theta_half)/theta_half

    scaled_rot_axis = (q.getVector()*frac_sin_theta_half).flatten()

    return cls(np.cos(theta_half), scaled_rot_axis[0], scaled_rot_axis[1], scaled_rot_axis[2])*np.exp(q.w)
```

<h2>Quaternion Logarithm:</h2>

<h4>Recap: Logarithm for Complex Numbers:</h4>
<p>
    given a complex number with magnitude
</p>

<p style="text-align: left; padding: 10px; font-size: 1.2em;">
    $$ z = re^{\theta i} = r(\cos(\theta) + i \sin(\theta)) $$
</p>

<p>
    the logarithm is defined as:
</p>

<p style="text-align: left; padding: 10px; font-size: 1.2em;">
    $$ \log(z) = \log(r) + \theta i  $$
</p>

<p>
    similarly the logarithmic mapping of a quaternion is defined:
</p>
<p style="text-align: left; padding: 10px; font-size: 1.2em;">
    $$ \log(q) = \log(\|q\|) + \frac{v}{\|v\|} \arccos\left(\frac{w}{\|q\|}\right) $$
</p>
Note:
<ul>
    <li>$\log(\|q\|) = 0$ for unit quaternions.</li>
    <li>zero-angle singularity when $ \|v\| = 0$ (can be handled with tailor series expansion).</li>
</ul>

</div>

#### Python code from Quaternion class:

```python
@classmethod
def log(self):
    norm_complex_vector = np.linalg.norm(self.getVector())

    theta_half = np.arctan2(norm_complex_vector, self.w)

    vector_scale = 0

    if norm_complex_vector < 0.001:
        vector_scale = (1.0 - (theta_half**2)/6.0 + (7.0*theta_half**4)/360.0)/self.norm()
    else:
        vector_scale = theta_half/norm_complex_vector

    vec = (self.getVector()*vector_scale).flatten()
    return Quaternion(np.log(np.linalg.norm(self.asVector())), vec[0], vec[1], vec[2])
```

<div style="font-family: Arial, sans-serif;">
<h2>Quaternions: Orientation and Direction</h2>
<p>Quaternions are a mathematical representation that can describe both rotation and orientation in 3D space. Unlike Euler angles, which represent rotations around individual axes, quaternions encapsulate a rotation about a vector in 3D space. This means that quaternions inherently carry information about both the magnitude (how much to rotate) and the direction (around which axis to rotate).</p>
<h3>Orientation vs. Rotation</h3>
<p>Orientation refers to a specific position in space, while rotation refers to the change in position. Euler angles describe only the rotation about each of its axes (pitch, yaw, and roll). However, when these rotations are combined in sequences, they can lead to a phenomenon called "gimbal lock," where one of the rotational degrees of freedom is lost. This can be problematic in applications like computer graphics or aerospace where continuous and smooth rotations are essential.</p>
<p>Quaternions, on the other hand, describe an orientation directly. This means that instead of describing how to get to a particular orientation through a series of rotations, quaternions simply describe the final orientation itself. This avoids the gimbal lock problem and provides a continuous and smooth representation of orientation.</p>
<h3>Why Euler Angles Can Be Problematic</h3>
<ul>
    <li><strong>Gimbal Lock:</strong> As mentioned, using Euler angles can lead to gimbal lock, a situation where you lose one degree of rotational freedom. This can be problematic in many applications where smooth and continuous rotations are required.</li>
    <li><strong>Interpolation Issues:</strong> Interpolating between two orientations using Euler angles can result in non-smooth transitions, especially near gimbal lock situations.</li>
    <li><strong>Redundancy:</strong> Multiple sets of Euler angles can represent the same orientation, leading to potential ambiguities.</li>
</ul>
<p>In conclusion, while Euler angles have their uses and can be intuitive in certain situations, quaternions offer a more robust and smooth representation for 3D rotations and orientations. However, as noted, they come with their own complexities and require a deeper understanding for effective application.</p>
</div>

<h1> Derivatives of the Quaternion exponential </h1>

given the pure quaternion $\bar{q} = 0 + v$ and writing the exponential funciton as $q = e^v = cos(\|v\|) + \frac{v}{\|v\|}\sin(\|v\|) = \cos(\sqrt{x^2 + y^2 + z^2}) + \frac{x \sin(\sqrt{x^2 + y^2 + z^2})}{\sqrt{x^2 + y^2 + z^2}}i + \frac{y \sin(\sqrt{x^2 + y^2 + z^2})}{\sqrt{x^2 + y^2 + z^2}}j+ \frac{z \sin(\sqrt{x^2 + y^2 + z^2})}{\sqrt{x^2 + y^2 + z^2}}k $ one can see that after the application of the chain rule $\dot{q} = \frac{\partial exp(v)}{\partial v}\frac{dv}{dt}$ the quaternion derivative can be calculated
    
for that we need the parial derivative $\frac{\partial exp(v)}{\partial v}$ which derives the exponential function by the vector components of v. These derivates are written below.

$$ \frac{\partial exp(\theta)_w}{\partial x} = -x\frac{\sin(\theta)}{\theta} $$
$$ \frac{\partial exp(\theta)_x}{\partial x} = x^2(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) + \frac{\sin(\theta)}{\theta} $$
$$ \frac{\partial exp(\theta)_y}{\partial x} = xy(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) $$
$$ \frac{\partial exp(\theta)_z}{\partial x} = xz(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) $$

<p> partial derivatives for y </p>

$$ \frac{\partial exp(\theta)_w}{\partial y} = -y\frac{\sin(\theta)}{\theta} $$
$$ \frac{\partial exp(\theta)_x}{\partial y} = yx(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3})$$
$$ \frac{\partial exp(\theta)_y}{\partial y} = y^2(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3})  + \frac{\sin(\theta)}{\theta} $$
$$ \frac{\partial exp(\theta)_z}{\partial y} = yz(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) $$

<p> partial derivatives for z </p>

$$ \frac{\partial exp(\theta)_w}{\partial z} = -z\frac{\sin(\theta)}{\theta} $$
$$ \frac{\partial exp(\theta)_x}{\partial z} = zx(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3})$$
$$ \frac{\partial exp(\theta)_y}{\partial z} = zy(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) $$
$$ \frac{\partial exp(\theta)_z}{\partial z} = z^2(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}) + \frac{\sin(\theta)}{\theta} $$



Given the quaternion exponential function with respect to the vector components \( x, y, z \), the Jacobian matrix \( J \) is:

$$
J = \begin{bmatrix}
\frac{\partial \exp(\theta)_w}{\partial x} & \frac{\partial \exp(\theta)_w}{\partial y} & \frac{\partial \exp(\theta)_w}{\partial z} \\
\frac{\partial \exp(\theta)_x}{\partial x} & \frac{\partial \exp(\theta)_x}{\partial y} & \frac{\partial \exp(\theta)_x}{\partial z} \\
\frac{\partial \exp(\theta)_y}{\partial x} & \frac{\partial \exp(\theta)_y}{\partial y} & \frac{\partial \exp(\theta)_y}{\partial z} \\
\frac{\partial \exp(\theta)_z}{\partial x} & \frac{\partial \exp(\theta)_z}{\partial y} & \frac{\partial \exp(\theta)_z}{\partial z} \\
\end{bmatrix}
$$

Substituting in the provided derivatives:

$$
J = \begin{bmatrix}
-x\frac{\sin(\theta)}{\theta} & -y\frac{\sin(\theta)}{\theta} & -z\frac{\sin(\theta)}{\theta} \\
x^2\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right)+ \frac{\sin(\theta)}{\theta} & yx\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) & zx\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) \\
xy\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) & y^2\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) + \frac{\sin(\theta)}{\theta} & zy\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) \\
xz\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) & yz\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) & z^2\left(\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}\right) + \frac{\sin(\theta)}{\theta} \\
\end{bmatrix}
$$

as seen, there occurs a new singularty in the $\frac{\cos(\theta)}{\theta^2} - \frac{\sin(\theta)}{\theta^3}$ term which needs to be resolved with a constant value for small $\theta$. To calculate the value we use the rule of L'Hospital:

<p>As $ \theta $ approaches 0, both terms become indeterminate forms of the type \( \frac{0}{0} \).</p>

<p>Using L'Hopital's rule for the first term:</p>
$$ \lim_{\theta \to 0} \frac{\cos(\theta)}{\theta^2} $$

<p>Differentiating the numerator with respect to $ \theta $ gives $ -\sin(\theta) $ and differentiating the denominator gives $ 2\theta $.</p>

<p>So, the limit becomes:</p>
\[ \lim_{\theta_{\text{half}} \to 0} \frac{-\sin(\theta_{\text{half}})}{2\theta_{\text{half}}} \]

<p>Again, this is of the form \( \frac{0}{0} \). Applying L'Hopital's rule again:</p>
\[ \lim_{\theta_{\text{half}} \to 0} \frac{-\cos(\theta_{\text{half}})}{2} = -\frac{1}{2} \]

<p>For the second term:</p>
\[ \lim_{\theta_{\text{half}} \to 0} \frac{\sin(\theta_{\text{half}})}{\theta_{\text{half}}^3} \]

<p>Differentiating the numerator gives \( \cos(\theta_{\text{half}}) \) and differentiating the denominator gives \( 3\theta_{\text{half}}^2 \).</p>

<p>So, the limit becomes:</p>
\[ \lim_{\theta_{\text{half}} \to 0} \frac{\cos(\theta_{\text{half}})}{3\theta_{\text{half}}^2} \]

<p>Again, this is of the form \( \frac{0}{0} \). Applying L'Hopital's rule twice more, we get:</p>
\[ \lim_{\theta_{\text{half}} \to 0} \frac{-\sin(\theta_{\text{half}})}{6\theta_{\text{half}}} \]
\[ \lim_{\theta_{\text{half}} \to 0} \frac{-\cos(\theta_{\text{half}})}{6} = -\frac{1}{6} \]

<p>Combining the results:</p>

<a id="screw-theory"></a>
<h1  style="font-size: 4.5em; border-bottom: 2px solid #ddd; padding-bottom: 10px;" >Screw Theory</h1>


Screw theory provides a mathematical framework that describes the spatial movement of rigid bodies. It's a powerful tool that combines rotational (angular) and translational (linear) motions into a single algebraic entity: the screw.

## Basics

- **Screw Axis**: Every motion in 3D space can be represented as a rotation about, and a translation along, a unique line in space called the screw axis.
- **Pitch**: The ratio of the translation to the rotation is called the pitch of the screw. If the pitch is infinite, the motion is purely rotational, and if the pitch is zero, the motion is purely translational.

## Plücker Coordinates

Screw motions and forces can be represented using Plücker coordinates, which are six numbers associated with lines in space. These coordinates are particularly useful for representing the kinematics and statics of rigid bodies.

## Applications

1. **Robotics**: Screw theory is used to describe robot kinematics, especially for serial and parallel manipulators.
2. **Mechanical Design**: It aids in understanding the constraints and motions in mechanisms.
3. **Computer Graphics**: Used for realistic animation of rigid body motions.

## Advantages

- **Unified Representation**: Screw theory combines rotation and translation, providing a compact representation of spatial displacements.
- **Algebraic Simplicity**: Many operations become algebraically simpler when expressed in terms of screws.
- **Intuitive**: Provides a geometrically intuitive way of understanding motion and forces.

---

## Forward Kinematics using Screw Theory

Forward kinematics refers to the computation of the end-effector position and orientation given the joint parameters (e.g., angles for revolute joints or displacements for prismatic joints).

### Skew-Symmetric Matrix

A skew-symmetric matrix is a square matrix that is the negative of its transpose. For a vector $ \mathbf{v} = [v_x, v_y, v_z]^T $, its skew-symmetric matrix is:

$$
[\mathbf{v}] =
\begin{bmatrix}
0 & -v_z & v_y \\
v_z & 0 & -v_x \\
-v_y & v_x & 0 \\
\end{bmatrix}
$$

The skew-symmetric matrix is useful in representing cross products in matrix form. For two vectors $ \mathbf{a} $ and $ \mathbf{b} $, the cross product $ \mathbf{a} \times \mathbf{b} $ can be represented as $ [\mathbf{a}] \mathbf{b} $.

### Rotation Matrices using Matrix Exponential

The matrix exponential offers a method to derive rotation matrices from algebraic representations. For a given rotation axis $ \hat{\mathbf{\omega}} $ and the angle $\theta$ that describes the rotation around the axis, the rotation matrix is:

$$ R(\hat{\mathbf{\omega}}, \theta) = e^{[\hat{\mathbf{\omega}}] \theta} $$

Recalling the matrix exponential:
$$ e^{At} = I + At + \frac{(At)^2}{2!} + \frac{(At)^3}{3!} + \dots $$

and the series expansion of $\sin(\theta)$ and $\cos(\theta)$:
$$ \sin(\theta) = \theta - \frac{\theta^3}{3!} + \frac{\theta^5}{5!} - \dots $$

$$ \cos(\theta) = 1 - \frac{\theta^2}{2!} + \frac{\theta^4}{4!} - \dots $$

The rotation matrix using the exponential coordinates $[\hat{\mathbf{\omega}}]\theta$ can be calculated as:


$$ e^{[\hat{\mathbf{\omega}}] \theta} = I + \sin(\theta) [\hat{\mathbf{\omega}}] + (1 - \cos(\theta)) [\hat{\mathbf{\omega}}]^2 $$

This is also known as the **Rodrigues' rotation formula**.

### Rigid-Body Motion using Matrix Exponential

Analogeously to the exponential Coordinates $[\hat{\mathbf{\omega}}]\theta$, the exponential coordinates of a homogenous transformation $T$ are defined as $\mathcal{S}\theta \in \mathbb{R}^6$, where $\mathcal{S} = (\hat{\mathbf{\omega}}, v)$ is the **screw axis** and $\theta$ is the distance travelled along the screw.

The screw axis can be represented as 4x4 matrix:

$$
[\mathcal{S}] =
\begin{bmatrix}
[\hat{\mathbf{\omega}}] & v  \\
0 & 0  \\
\end{bmatrix}
$$

with $[\mathcal{S}]\theta$ the transformation can then be written as:

$$ T(\mathcal{S}, \theta) = e^{[\mathcal{S}] \theta} = \begin{bmatrix}
e^{[\hat{\mathbf{\omega}}] \theta} & (I\theta + (1 - \cos\theta)[\hat{\mathbf{\omega}}] + (\theta - \sin\theta)[\hat{\mathbf{\omega}}]^2)v  \\
0 & 0  \\
\end{bmatrix}$$


### Forward Kinematics

For a robot with a series of joint motions:

1. **Define the Screw Axis**: For each joint, define the screw axis $ \mathcal{S} $ that represents the motion of that joint. This screw axis will be a 6x1 vector
   
2. **Compute the Matrix Exponential**: For each joint motion $ \mathcal{S}_i $ and its associated joint parameter $ \theta_i $, compute the matrix exponential $ e^{[\mathcal{S}_i] \theta_i} $. This results in a transformation matrix for each joint.
   
3. **Multiply the Transformations**: The overall transformation from the robot's base to its end-effector is the product of the individual transformations. If there are $ n $ joints:

$$
\mathbf{T}_{\text{base}}^{\text{end-effector}} = e^{[\mathcal{S}_1] \theta_1} e^{[\mathcal{S}_2] \theta_2} ... e^{[\mathcal{S}_n] \theta_n}M
$$

This transformation matrix $ \mathbf{T}_{\text{base}}^{\text{end-effector}} $ provides the position and orientation of the end-effector given the joint parameters and the Endeffktor position in the robots home position $M$.


# Robot Kinematics: DH Parameters vs. Screw Theory vs. Lie Algebra

## Generality:
- **DH Parameters**: Designed for serial link manipulators with single DOF joints. Limited to certain robot structures.
- **Screw Theory**: Applicable to both serial and parallel manipulators. Describes translational and rotational motion.
- **Lie Algebra**: Describes continuous transformations. Suitable for various robotic systems.

## Compactness and Intuitiveness:
- **DH Parameters**: Compact for serial manipulators but may lack intuitiveness.
- **Screw Theory**: Intuitive representation of motion and forces.
- **Lie Algebra**: Compact representation, though may be less intuitive than screw theory.

## Avoidance of Singularities:
- **DH Parameters**: Can lead to kinematic singularities.
- **Screw Theory**: Addresses singularities effectively.
- **Lie Algebra**: Provides insights into motion near singular configurations.

## Interpolation and Motion Planning:
- **DH Parameters**: Challenges in interpolation, especially for orientation.
- **Screw Theory**: Natural interpolation between poses.
- **Lie Algebra**: Suitable for smooth interpolation and trajectory generation.

## Extensibility:
- **DH Parameters**: Limited extensibility for new configurations or DOFs.
- **Screw Theory**: Highly extensible across robotic configurations.
- **Lie Algebra**: General framework, extensible to various systems.

**Conclusion**: While DH parameters are foundational in robot kinematics, Screw Theory and Lie Algebra offer broader applicability and insights. The optimal choice depends on the problem and user familiarity.

<h1  style="font-size: 4.5em; border-bottom: 2px solid #ddd; padding-bottom: 10px;" >Introduction to Dual Quaternions</h1>

Dual quaternions are an extension of quaternions that are used primarily for representing rigid transformations, including rotation and translation in 3D space. They are especially popular in computer graphics, robotics, and kinematics.

## Definition of unit Dual Quaternions

A dual quaternion is composed of two quaternions: a unit quaternion $ q_r $  which represents the orientation and rotation of the Transformation and a quaternion $ q_d $. $ q_d $ represents the translation of the Transformation and is not neccessarily unit length.

$$ \underline{\xi} = q_r + \epsilon q_d $$

where $ \epsilon $ is a dual unit with the property $ \epsilon^2 = 0 $.

<h4>Note:</h4>
<ul>
    <li>$q_r$ is also called the real part of the Dual Quaternion</li>
    <li>$q_d$ is also called the dual part of the Dual Quaternion.</li>
</ul>


## Definition of pure Dual Quaternions - Dual Vectors

similarly to quaternions we can use dual quaternions to describe two vectors $x$ and $y$ in $\mathbb{R}^3$

$$ 
\underline{\xi} = (0, x_1, x_2, x_3) + \epsilon (0, y_1, y_2, y_3)
$$

<h4>These can be used to represent:</h4>
<ul>
    <li>wrenches (moment, force)</li>
    <li>twists (angular velocity, linear velocity).</li>
    <li>accelerations (angular acceleration, linear acceleration).</li>
    <li>screw axis (rotation axis, moment of line).</li>
</ul>

## Advantages over Matrices

- **Compact Representation**: Dual quaternions require only 8 scalar values, compared to 16 for 4x4 matrices.
- **Smooth Interpolation**: Dual quaternions allow for smooth interpolation between transformations using techniques like Screw Linear Interpolation (SCLERP).
- **Contained Infromation**: Dual quaternions contain besides the Orientation information also the direction of rotation, and are **singularity free**.
- **Conversion**: Dual quaternions can be converted to Transformation matrices any time (under loss of direction information). Backwards conversion is bad because the information of direction in the Rotationmatrix is random when converted to a quaternion
- **Easier Algebraic Operations**: Line Transformations are possible with very little effort, allowing for elegant solutions



## Multiplication of Dual Quaternions and the Dual Unit

The multiplication of two dual quaternions $\underline{\xi}_1$ and $\underline{\xi}_2$ can be done as follows:

$$
\underline{\xi}_1 \otimes \underline{\xi}_1 = (q_r^{(1)} + \epsilon q_d^{(1)}) \otimes (q_r^{(2)} + \epsilon q_d^{(2)})
$$

$$
\underline{\xi}_1 \otimes \underline{\xi}_1 = (q_r^{(1)} \otimes q_r^{(2)}) + \epsilon(q_r^{(1)} \otimes q_d^{(2)} + q_d^{(1)} \otimes q_r^{(2)}) + \epsilon^2(q_d^{(1)} \otimes q_d^{(2)})
$$

with $ \epsilon^2 = 0 $ the new dual quaternion is written like 

$$
\underline{\xi}_3= (q_r^{(1)} \otimes q_r^{(2)}) + \epsilon(q_r^{(1)} \otimes q_d^{(2)} + q_d^{(1)} \otimes q_r^{(2)})
$$



Please note that quaternion multiplication should follow the rules of quaternion algebra, including the Hamilton product rule for quaternion multiplication.

### Comparison to Homogeneous Transfromation Matrices:

Dual Quaternion Multiplication is similar to the multiplication of Transformation matrices 

$$
\underline{\xi}_2^0= \underline{\xi}_1^0 \otimes \underline{\xi}_2^1 
$$

corresponds to: 



$$
T_2^0= T_1^0 T_2^1 
$$

## Screw axis Forward Kinematics with Dual Quaternions
with the position vectors and rotation axis we can now construct the screw axis in dualquaternion form.

with this the screws $ \underline{s} = r + \epsilon m = r + \epsilon (r \times p) $ are internally defined.

with the screws and the angle we can now define the dual quaternion transformations $\underline{\xi}_i$ which correspond to motion around each robotic joint

$$
\underline{\xi}_i = e^{\frac{\underline{\theta}_i}{2}\underline{s}_i} 
$$

with these transformations we can then define the transformation between the baseframe and the tcp $ \underline{\xi}_{TCP}^0 $

$$ 
\underline{\xi}_{TCP}^0 = (\Pi_{i = 0}^{n = 7} \underline{\xi}_i) M
$$

````python
from DualQuaternions.DualQuaternion import DualQuaternion
from DualQuaternions.Quaternion import Quaternion

s1 = DualQuaternion.screwAxis(*r1, *p1)
s2 = DualQuaternion.screwAxis(*r2, *p2)
s3 = DualQuaternion.screwAxis(*r3, *p3)
s4 = DualQuaternion.screwAxis(*r4, *p4)
s5 = DualQuaternion.screwAxis(*r5, *p5)
s6 = DualQuaternion.screwAxis(*r6, *p6)
s7 = DualQuaternion.screwAxis(*r7, *p7)

M = DualQuaternion.fromQuatPos(Quaternion(1,0,0,0), [0, 0, d1 + d3 + d5 + d7])

def forward_kinematics(p):
    x1 = DualQuaternion.fromScrewAxis(p[0], 0, s1)
    x2 = DualQuaternion.fromScrewAxis(p[1], 0, s2)
    x3 = DualQuaternion.fromScrewAxis(p[2], 0, s3)
    x4 = DualQuaternion.fromScrewAxis(p[3], 0, s4)
    x5 = DualQuaternion.fromScrewAxis(p[4], 0, s5)
    x6 = DualQuaternion.fromScrewAxis(p[5], 0, s6)
    x7 = DualQuaternion.fromScrewAxis(p[6], 0, s7)
    
    return x1*x2*x3*x4*x5*x6*x7*M
````

<h2>Steps for Kinematic Calibration with a Laser Tracker:</h2>

<ol>
    
<li><strong>Initial Measurements:</strong> Move the robot to several predefined configurations and measure the TCP position with the Laser Tracker. </li>

<li><strong>Linearize Forward Kinematics:</strong> Using the robot's kinematic model, compute the expected position and orientation of the end-effector for each of the predefined positions. Compare these expected values with the measurements from the laser tracker to determine the errors.</li>
    
<li><strong>Modeling Errors:</strong> Using the robot's kinematic model, compute the expected position and orientation of the end-effector for each of the predefined positions. Compare these expected values with the measurements from the laser tracker to determine the errors.</li>

<li><strong>Parameter Identification:</strong> Using optimization techniques, adjust the parameters of the robot's kinematic model to minimize the discrepancies between the expected and measured end-effector positions and orientations. This might involve adjusting link lengths, joint offsets, or other kinematic parameters.</li>

<li><strong>Validation:</strong> Once the parameters have been identified and adjusted, move the robot to new positions that were not used in the initial measurements. Measure the end-effector's position and orientation with the laser tracker and compare these measurements to the predictions from the updated kinematic model to validate the calibration.</li>

<li><strong>Iterative Refinement:</strong> If the validation step shows discrepancies that are larger than acceptable thresholds, repeat the calibration process with more positions or refine the kinematic model further.</li>

<li><strong>Finalization:</strong> Once the calibration is validated and the discrepancies are within acceptable limits, update the robot's control system with the new kinematic parameters to ensure accurate motion planning and control.</li>
</ol>

## Calculation and Error Evaluation

This results in an error for pose $i$ 

$$
p_{error}^{i} = p_{meas}^i - p_{FK}^i = \begin{bmatrix}
x_{meas}^{i} - x_{FK}^{i} \\
y_{meas}^{i} - y_{FK}^{i} \\
z_{meas}^{i} - z_{FK}^{i}
\end{bmatrix}
$$


$$
p_{error}^{i} \approx \delta p^i
$$
This calculation is done for all poses \(n\) of the set of configurations. The parameter identification process is based on minimizing the error of the end-effector position. The robot kinematics are mildly nonlinear, but linearization is an effective way for iteratively solving a nonlinear optimization. Therefore, the Jacobian matrix is used [[3], [18]]. The Jacobian is calculated for every single pose \(i\) and every kinematic parameter \(l\). The Jacobian matrix \(J^i\) for pose \(i\) describes the change of the kinematic pose \( \delta x^i \) (here position \(x, y,\) and \(z\)) of the end-effector in dependence of the change of the kinematic parameters \( \delta p \). Respectively, it represents the relationship between the parameter errors and the end-effector position [[10], [12]].

$$
\delta x^i = J^i \delta p
$$


The Jacobian $J^i$ is then:

$$
J^i = \begin{bmatrix}
\frac{\partial x^i}{\partial p_{nom}} & \frac{\partial y^i}{\partial p_{nom}} & \frac{\partial z^i}{\partial p_{nom}}
\end{bmatrix}^T
$$

Note: 

the rotation vector parameters can not be derived directly, it would probably lose unit lenght.

to solve this isse we use a trick and rotate the vector with a rotation matrix and use the 3 euler angles to derive that rotation axis of the Screw axis.

derivation of the position can be done directly


````python
from DualQuaternions.DualQuaternion import DualQuaternion
from DualQuaternions.Quaternion import Quaternion
import numpy as np

def Rot_rpy(roll, pitch, yaw):
    
    # Roll matrix
    Rx = np.array([[1, 0, 0],
                    [0, np.cos(roll), -np.sin(roll)],
                    [0, np.sin(roll), np.cos(roll)]])
    
    # Pitch matrix
    Ry = np.array([[np.cos(pitch), 0, np.sin(pitch)],
                    [0, 1, 0],
                    [-np.sin(pitch), 0, np.cos(pitch)]])
    
    # Yaw matrix
    Rz = np.array([[np.cos(yaw), -np.sin(yaw), 0],
                    [np.sin(yaw), np.cos(yaw), 0],
                    [0, 0, 1]])
    
    # Combined rotation matrix
    R = Rz@Ry@Rx
    return R


def forward_kinematics(q, p):
    z = [0,0,1]
    y = [0,1,0]
    
    r1 = Rot_rpy(p[1], p[2], p[3]) @z
    r2 = Rot_rpy(p[8], p[9], p[10])@y
    r3 = Rot_rpy(p[15], p[16], p[17])@z
    r4 = Rot_rpy(p[22], p[23], p[24])@y
    r5 = Rot_rpy(p[29], p[30], p[31])@z
    r6 = Rot_rpy(p[36], p[37], p[38])@y
    r7 = Rot_rpy(p[43], p[44], p[45])@z
    
    x1 = DualQuaternion.fromScrewAxis(q[0] + p[0],  0, DualQuaternion.screwAxis(*r1, p[4],  p[5],  p[6]))
    x2 = DualQuaternion.fromScrewAxis(q[1] + p[7],  0, DualQuaternion.screwAxis(*r2, p[11], p[12], p[13]))
    x3 = DualQuaternion.fromScrewAxis(q[2] + p[14], 0, DualQuaternion.screwAxis(*r3, p[18], p[19], p[20]))
    x4 = DualQuaternion.fromScrewAxis(q[3] + p[21], 0, DualQuaternion.screwAxis(*r4, p[25], p[26], p[27]))
    x5 = DualQuaternion.fromScrewAxis(q[4] + p[28], 0, DualQuaternion.screwAxis(*r5, p[32], p[33], p[34]))
    x6 = DualQuaternion.fromScrewAxis(q[5] + p[35], 0, DualQuaternion.screwAxis(*r6, p[39], p[40], p[41]))
    x7 = DualQuaternion.fromScrewAxis(q[6] + p[42], 0, DualQuaternion.screwAxis(*r7, p[46], p[47], p[48]))
    M = DualQuaternion.fromQuatPos(Quaternion(1,0,0,0), [p[49], p[50], p[51]])
    return x1*x2*x3*x4*x5*x6*x7*M

q = [0,1,1,1,0,0,0]

d1 = 0.438
d3 = d1 + 0.7
d5 = d3 + 0.7
d7 = d5 + 0.115

p0 = [0,0,0,0,0,0,d1,0,0,0,0,0 ,0 ,d1, 0, 0, 0, 0, 0, 0,d3, 0, 0, 0, 0, 0, 0,d3, 0, 0, 0, 0, 0, 0,d5, 0, 0 ,0, 0, 0, 0,d5, 0, 0, 0, 0, 0, 0,d5, 0, 0,d7]
T = forward_kinematics(q, p0)

print(T)
    
````

## Forward Kinematics Derivative

The Jacobian for pose $ Q_i $ of $n$ poses and kinematic parameter $p_l $ of $n_p$ parameters is calculated numerically with a vectorial central difference quotient. This is shown in the following equations:

To the parameter $p_l$ in the parameter vector the small error $ h = 0.0001$, is added to get the vector $p_{+}$ and subtracted to get the vector $p_{-}$:

$$
p_{+}^k = \begin{bmatrix}
p_1 & p_2 & \dots & p_k + h & \dots & p_{n}
\end{bmatrix}^T
$$


$$
p_{-}^k = \begin{bmatrix}
p_1 & p_2 & \dots & p_k - h & \dots & p_{n}
\end{bmatrix}^T
$$

with those 2 parameter vectors we can take the simple derivative

$$ J_k = \frac{dFK(q,p)}{dp_k} \approx \frac{FK(q,p_+^k) - FK(q,p_-^k)}{2h}$$

````python
def fk_derivative(q, p, num_param):
    h = 0.001
    p_pos = p.copy()
    p_neg = p.copy()
    
    p_pos[num_param] += h
    p_neg[num_param] -= h
    
    T_1 = forward_kinematics(q, p_pos)
    T_2 = forward_kinematics(q, p_neg)
    
    d_pos = (T_1.getPosition() - T_2.getPosition())/(2*h)
    return d_pos
````

to get the full jacobian for a single pose we iterate over all parameters and calculate the derivative of the parametrized Forward Kinematics

````python
def get_single_jacobian(q,p):
    
    J = np.zeros((3, len(p)))
    for i in range(len(p)):
        J[:, i] = fk_derivative(q,p,i).flatten()
    
    return J

np.set_printoptions(precision=2, suppress=True, linewidth=200, formatter={'float': '{:8.3f}'.format})
#print(get_single_jacobian(q,p0))
````

to get the full jacobian for the whole dataset we iterate over all Joint configurations and poses and compose the final matrix of the parameter submatrices

````python
def get_Jacobian(Q, p):
    J = np.zeros((3*len(Q), len(p)))
    for i in range(len(Q)):
        J[3*i:3*i+3, :] = get_single_jacobian(Q[i], p)
        
    return J

#print(get_Jacobian(Q_list, p0))
````

# Solving Linear Systems

Linear systems of the form $b = Jx$ can be solved by inversion of $J$ if the jacobian is square, which leads to $x = J^{-1}b$. But what to do if $J$ is not square?

## Pseudo-Inverse Solution

The solution to the linear system $b = Jx$ using the pseudo-inverse is derived by left hand side extension of the formula by $J^T$:


$$
J^T b = J^TJ x
$$

$J^TJ$ is now square and thus should be invertible, which leads the the left-hand side pseudo-inverse of J:

$$
x = (J^TJ)^{-1}J^Tb
$$

Where:
- $x$ is the solution vector.
- $A^+$ is the pseudo-inverse of matrix \(A\).
- $b$ is the right-hand side vector.


### Code example:

````python
import numpy as np

J = get_Jacobian(Q, p0)

J_pinv = np.linalg.inv(J.T @ J) @ J.T

print(J_pinv)
````

## Having a deeper look

A good tool to see what is going on is the singular value decomposition (SVD) of matrix $J$:

$$
J = U\Sigma V^T
$$

Where:
- $U$ is the left singular matrix.
- $\Sigma$ is a diagonal matrix containing the singular values of $J$.
- $V^T$ is the transpose of the right singular matrix.

````python
import numpy as np

# Compute the SVD of the Jacobian matrix
U, S, VT = np.linalg.svd(J)

# Calculate the condition number
condition_number = np.max(S) / np.min(S)

# Print the singular values and condition number
print("Singular Values of J:")
print(S)

print("Condition Number of J:")
print(condition_number)
````

## Handling Rank-Deficient Matrices

If matrix $A$ is rank-deficient (i.e., it has fewer linearly independent rows or columns than its size), the pseudo-inverse approach can still be used with some considerations:

1. **Adjust Problem Setup:** Remove linear dependent Parameters s.t. that the Jacobian has full rank.

2. **Regularization:** To handle rank deficiency, you can add a small positive value $\lambda$ to the diagonal of $J^TJ$ to make it invertible.

$$
x = (J^TJ + \lambda I)^{-1}J^Tb
$$



    Analogeuosly we can calculate the same damped pseudo inverse with the singular value decomposition:



$$
x = V \Sigma^\dagger U^Tb
$$

    with 
    
$$
\Sigma^\dagger = diag(\frac{1}{\sigma_i + \lambda})
$$

### Code example:

````python
import numpy as np

u = 0.0001

J = get_Jacobian(Q, p0)

I = np.eye(J.shape[1])
J_damped_pinv = np.linalg.inv(J.T @ J + u*I) @ J.T

print(J_damped_pinv)
````

## Calibration 

to find the correct set of parameters for the robot we will use an iterative process:

here we calculate the positions of the FK with the current parameters and the given set of joint configurations

$$
p_{FK}^i = FK(Q_i, p_k)
$$

with the measured position from the real robot and the stacked vector of $ p_{FK} $ we calculate the error, which approximates the gradient of $ dp = Jdp $.

and with the pseudo inverse we can then iteratively update the parameter vector as such:

$$
p_{k+1} = p_{k} + J^\dagger e
$$

````python
p = p0.copy()

u = 0.0001

max_iter = 100

for k in range(max_iter):
    J = get_Jacobian(Q, p)
    
    position_model_list = []
    for i in range(len(Q)):
        position_model_list.append(forward_kinematics(Q[i], p).getPosition().flatten())
        
    #print(position_model_list)
    position_model_vector = np.vstack(position_model_list).flatten()
    #print(position_model_vector)
    error = position_real_vector - position_model_vector
    
    error_norm = np.linalg.norm(error)
    
    print(error_norm)
    if error_norm < 1e-8:
        break

    I = np.eye(J.shape[1])
    J_inv = np.linalg.inv(J.T @ J + u * I) @ J.T
    dp = J_inv@error
    
    p += dp
#print(p0)
print(p)
    
````

<h2>Steps for Kinematic Calibration with a Laser Tracker:</h2>

<ol>
<li><strong>Setup:</strong> Position the laser tracker such that it has a clear line of sight to the robot's end-effector and the robot's workspace. Ensure that the robot and the laser tracker are properly synchronized.</li>

<li><strong>Finalization:</strong> Once the calibration is validated and the discrepancies are within acceptable limits, update the robot's control system with the new kinematic parameters to ensure accurate motion planning and control.</li>
</ol>
